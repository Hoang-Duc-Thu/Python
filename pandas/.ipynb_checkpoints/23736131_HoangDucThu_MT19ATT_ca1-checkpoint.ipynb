## Phần 1: Numpy

```python
import numpy as np
import pandas as pd
```

```python
#1. Tạo mảng ngẫu nhiên arr1 gồm 10 phần tử số nguyên từ 10-29
np.random.seed(2)
arr1 = np.random.randint(10,30,size=10)
arr1
```

```python
#2. Tạo mảng arr2 là mảng arr1 nhưng có các giá trị chia hết cho 3 được thay bằng -1
arr2 = np.where(arr1 % 3==0, -1, arr1)
arr2
```

```python
#3. Tạo mảng arr3 chứa 10 phần tử với các phần tử ở index lẻ là 0 và ở index chẵn là số chẵn.
arr3 = np.arange(10)
arr3[arr3 % 2 != 0] = 0
arr3
```

```python
#4.  Liệt kê ra các phần tử giống nhau trong arr1 và arr3 (bang value va index)
l = []
for i in range(len(arr1)):
    if arr1[i] == arr3[i]:
        l.append((i, arr1[i]))
l
```

```python
#5. Phát sinh mảng 2 chiều arr_2D (4x6) với mỗi phần tử ngẫu nhiên từ 1->100 
np.random.seed(3)
arr_2D = np.random.randint(1,101,size=24)
arr_2D = arr_2D.reshape((4,6))
arr_2D
```

```python
#6. Liệt kê các dòng trong arr_2D mà phần tử cột cuối là chẵn 
arr_2D[arr_2D[:,-1]%2 == 0]
```

```python
#7. Tính trung bình các dòng lẻ
np.mean(arr_2D[1::2],axis=1)
```

```python
#8. Thêm cột 7 là tổng của mỗi dòng trong arr_2D
new_row = np.sum(arr_2D,axis= 1).reshape(-1,1)
arr_2D = np.hstack((arr_2D,new_row))
arr_2D
```

```python
#9. Thêm dòng thứ 5 là tổng của mỗi cột trong arr_2D
new_col = np.sum(arr_2D,axis= 0).reshape(1,-1)
arr_2D = np.vstack((arr_2D,new_col))
arr_2D
```

```python
#10. Liệt kê các phần tử trong mảng arr_2D là số nguyên tố.
def nguyento(n):
    if n < 2:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True
arr_2D[np.vectorize(nguyento)(arr_2D)]
```

## Phần 2: Pandas

```python
#1. Đọc file dữ liệu đã cho shoefly.csv
df = pd.DataFrame(pd.read_csv('shoefly.csv',sep=','))
df
```

```python
#2. Thêm 1 cột total = price * amount
df['total'] = df['price'] * df['amount']
df
```

```python
#4. Cho biết Frances Palmer đã đặt hàng gì?
df.loc[df['first_name'] +' '+ df['last_name'] == 'Frances Palmer']
```

```python
#5. Hãy liệt kê tất cả các đơn đặt hàng có loại giày là clogs, boots, và ballet flats
df[(df['shoe_type'] == 'clogs')|(df['shoe_type'] == 'boots')|(df['shoe_type'] == 'ballet flats')  ]
```

```python
#6. Hãy thêm một cột mới shoe_source, nếu các vật liệu là leather thì là animal, ngược lại thì ghi là vegan  
df['shoe_source'] = df['shoe_material'].apply(lambda x: 'animal' if x == 'leather' else 'vegan')
df
```

```python
#7. Hãy cho biết giá trị đơn hàng trung bình của các mặt hàng giày thuộc nhóm animal là bao nhiêu
df[df['shoe_source'] == 'animal']['total'].mean()
```

```python
#8. Hiển thị orders theo thứ tự giảm dần orders và tăng dần last_name
df.loc[:,['order_id','last_name']].sort_values(by=['order_id','last_name'] , ascending=[False,True])
```

```python
#9 Thống kê số lượng khách nam và nữ.
df['gender'].value_counts().reset_index()
```

```python
#10 Hãy cho biết giá trung bình của mỗi loại giày.
df.groupby('shoe_type')['total'].mean().reset_index().rename(columns={'total': 'mean'})
```
