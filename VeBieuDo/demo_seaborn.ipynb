# Tạo grid trong seaborn

# FacetGrid

```python
# Bước 1: tạo lươi
g = sb.FacetGrid(data=df,col='day')
#Bước 2: Ánh xạ biểu đồ vào ô lưới
g.map(sb.scatterplot,'total_bill','tip')
```

### 4 hàm không map vào ô lưới trong FacetGrid : catplot, displot, lmplot, relplot

```python
sb.catplot(data=df,col='day',kind='count',x='sex',col_wrap=3)
```

```python
sb.displot(data=df,x='total_bill',col='day',hue='sex',kind='kde')
```

```python
sb.relplot(data=df,x='total_bill',y='tip',col='day',hue='sex',col_wrap=2)
```

```python
sb.scatterplot(data=df,x='total_bill',y='tip')
```

```python
sb.lmplot(data=df,x='total_bill',y='tip',col='day',hue='sex',col_wrap=2)
```

# PairGrid

```python
# Bước 1: tạo lươi
g = sb.PairGrid(data=df[['total_bill','tip','sex']],hue='sex')
#Bước 2: Ánh xạ biểu đồ vào ô lưới
g.map_diag(sb.histplot)
g.map_offdiag(sb.scatterplot)
```

```python
# Bước 1: tạo lươi
g = sb.PairGrid(data=df[['total_bill','tip','sex']],hue='sex')
#Bước 2: Ánh xạ biểu đồ vào ô lưới
g.map_diag(sb.histplot)
g.map_upper(sb.scatterplot)
g.map_lower(sb.regplot)
```

```python
# shortcut pairgrid
```

```python
sb.pairplot(data=df[['total_bill','tip','sex']],hue='sex')
```

# JoinGrid

```python
# Bước 1: tạo lươi
g = sb.JointGrid(data=df,x='total_bill',y='tip',hue='sex')
#Bước 2: Ánh xạ biểu đồ vào ô lưới
g.plot(sb.scatterplot,sb.histplot)
```

```python
# shortcut
```

```python
sb.jointplot(data=df,x='total_bill',y='tip',kind='reg')
```
