```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

# Vẽ trên tầng scriptting layer

```python
x = np.linspace(1,10,1000)
y = np.sin(x)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
plt.title('Biểu đồ sin(x)',color='#0F0')
plt.xlabel('X',color='r',fontsize=14)
plt.ylabel('sin(x)',fontsize=14)
plt.show()
```

# Vẽ trên tầng artist layer

```python
fig,ax = plt.subplots(nrows=1,ncols=1,figsize=(10,5))
ax.plot(x,y)
ax.set_title('Biểu đồ sin(x)')
ax.set_xlabel('X',color='r',fontsize=14)
ax.set_ylabel('sin(x)',fontsize=14)
plt.show()
```

```python
fig,ax = plt.subplots(nrows=2,ncols=2,figsize=(10,5))
ax[0,0].plot(x,y)
ax[0,0].set_title('Biểu đồ sin(x)')
ax[0,0].set_xlabel('X',color='r',fontsize=14)
ax[0,0].set_ylabel('sin(x)',fontsize=14)
plt.tight_layout()
plt.show()
```

```python
b = np.random.randint(1,20,size=10)
plt.plot(b,'d-.c')
plt.title('Biểu đồ sin(x)',color='#0F0')
plt.xlabel('X',color='r',fontsize=14)
plt.ylabel('sin(x)',fontsize=14)
plt.show()
```

# Biểu đồ bar

```python
#ten = ['an','binh','cuc','dung','tuan']
stt = np.arange(5)
diem = [5,8,6,9,7]
```

```python
plt.bar(x=stt,height=diem,color=['r','b','g','y','c'])
plt.xticks(stt,['an','binh','cuc','dung','tuan'],rotation=40)
plt.ylim(0,10)
plt.show()
```

```python
plt.barh(y=stt,width=diem,color=['r','b','g','y','c'])
plt.yticks(stt,['an','binh','cuc','dung','tuan'],rotation=40)
plt.xlim(0,10)
plt.show()
```

# Biểu đồ pie
